# Baseline training

The goal of this part is to explore 2 detectors: YOLOv8 and Faster R-CNN.

## Import libraries and function

```python
from pathlib import Path
import sys, os
sys.path.append(os.path.abspath("../src"))

from sklearn.model_selection import train_test_split
import shutil
import yaml
import warnings
import torch

from cadot.utils.path import get_data_path, add_src_to_sys_path
add_src_to_sys_path()
from cadot.data.loading import get_image_label_pairs
from cadot.models.yolo import CADOT_CLASS_NAMES, prepare_yolo_dataset, train_yolo
```

## Train YOLO baseline

```python
data_root = get_data_path()

out_dir = Path("yolo_dataset_split")

# prepare the dataset 
data_yaml = prepare_yolo_dataset(out_dir=out_dir, class_names=CADOT_CLASS_NAMES)
```

```python
# train the model
print(f"\nMPS available: ")
print(torch.backends.mps.is_available())
print(f"\nGPU available: ")
print(torch.cuda.is_available())

if torch.backends.mps.is_available():
    train_yolo(
        data=str(data_yaml),
        model="yolov8n.pt",
        epochs=5,
        imgsz=512,
        batch=8,
        device="mps", # config mac
        val=False
    )
else:
    train_yolo(
        data=str(data_yaml),
        model="yolov8n.pt",
        epochs=10,
        imgsz=512,
        batch=8,
        device="0", # NVIDIA GPU
        val=False
    )
```

### See results on the validation set

This is relevant only if we uses `val=False` during the training. In case `val=True`, use the test set (copy not created yet).

```python
from ultralytics import YOLO
model = YOLO("runs/train/train4/weights/best.pt")
metrics = model.val(data=str(data_yaml))
model.predict(source="yolo_dataset_split/val/images", save=True)
```
